### Loading the Dataset

```python
# Dataset source URL
url = "https://raw.githubusercontent.com/data-bootcamp-v4/prework_data/main/students_performance.csv"
```

```python
import seaborn as sns
import pandas as pd
sp = pd.read_csv(url)
```

### Exploring Dataset

```python
# Number of rows and columns
sp.shape
```

```python
# Number of unique values for each column
sp.nunique()
```

### Visualizing Distributions

### Histograms for math, reading, and writing scores

```python
sns.histplot(data=sp, x="math score")
```

```python
sns.histplot(data=sp, x="reading score")
```

```python
sns.histplot(data=sp, x="writing score")
```

##### Numerical variables, interpretation of shape distribution
* Math scores: the majority of students are in the range of approximately 50 to 80. Very few are below 40 or over 90, with some outliers getting closer to 0.
* Reading scores: the majority of students are in the range of approximately 55 to 85, with a pick in the range around 70 to 75. Fewer students below 50 or over 90.
* Writing scores: the majority of students are in the range of approximately 55 to 85, with a pick in the range 70 to 75. Fewer students below 50 or over 90.

### Bar plots for categorical variables

```python
sns.countplot(data=sp, x="gender")
```

```python
sns.countplot(data=sp, x="race/ethnicity")
```

```python
sns.countplot(data=sp, x="parental level of education")
```

```python
sns.countplot(data=sp, x="lunch")
```

```python
sns.countplot(data=sp, x="test preparation course")
```

##### Categorical variables, interpretation of shape distribution
* Gender: The bar plot shows a similar amount of male and female students.
* Race / Ethnicity: Significant differences between the groups, with the majority of students being part of groups C and D, a second less represented cohort being groups B and E, and a minoritary group A.
* Parental level of education: Main group in between "some high school" and "associate's degree", with the minority being bachelor's and, finally, master's.
* Lunch: double the amount of students receive standard lunch, compared to the students receiving free/reduced lunch.
* Test preparation course: double the amount of students didn't take the test preparation course, compared to the students who did. There's a correlation with the lunch variable.
