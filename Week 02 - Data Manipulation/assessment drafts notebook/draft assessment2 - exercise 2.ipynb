```python
import pandas as pd
att = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/attention.csv")
att.drop(columns=["Unnamed: 0"], inplace=True)
att
```

```python
att_focus = att[att["attention"] == "focused"].pivot_table(index="subject", columns=["solutions"], values=["score"])
```

```python
att_div = att[att["attention"] == "divided"].pivot_table(index="subject", columns=["solutions"], values=["score"])
```

```python
att.pivot_table(index="subject", columns=["solutions"], values=["score"])
```

```python
focus_avg = att[att["attention"] == "focused"]["score"].mean()
```

```python
div_avg = att[att["attention"] == "divided"]["score"].mean()
```

```python
focus_avg / div_avg
```

```python
solution_avgs = {
      1 : att[att["solutions"] == 1]["score"].mean(),
      2 : att[att["solutions"] == 2]["score"].mean(),
      3 : att[att["solutions"] == 3]["score"].mean()
    }

max(solution_avgs, key=solution_avgs.get)
```

```python
sub_avgs = pd.DataFrame(att.groupby(["subject"])["score"].mean())
```

```python
sub_avgs = sub_avgs.sort_values(by=["score", "subject"], ascending=[False, True])[:10].reset_index()
```
