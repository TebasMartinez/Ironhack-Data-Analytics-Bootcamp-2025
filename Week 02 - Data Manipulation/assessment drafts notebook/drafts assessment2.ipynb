```python
import pandas as pd
pop = pd.read_csv("data/c2119.csv")
obesity = pd.read_csv("data/c2228.csv")
```

```python
pop_obty = pd.merge(pop, obesity, on="Name", how="outer")
pop_obty.columns = ["Country", "Population", "Obesity Rate"]
```

```python
pop_obty
```

```python
pop_obty.dropna(how="any")
```

```python
pop_obty_higher20_largepop = pop_obty[(pop_obty["Population"] > 10 ** 7) & (pop_obty["Obesity Rate"] >20)]
pop_obty_higher20_largepop
```

```python
result = pop_obty_higher20_largepop.sort_values(by=["Population"], ascending=False).sort_values(by=["Obesity Rate"], ascending=False)
```

```python
result = result.reset_index(drop=True)
result.index += 1
```
